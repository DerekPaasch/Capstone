# Cleaning Healthcare_coverage_clean

```python
Healthcare_coverage_clean = Healthcare_coverage.drop(columns = ['STRUCTURE', 'STRUCTURE_ID', 'STRUCTURE_NAME', 'ACTION', 'REF_AREA', 'FREQ', 'MEASURE', 'Time period', 'Observation value', 'OBS_STATUS', 'DECIMALS', 'Decimals'])
Healthcare_coverage_clean.reset_index(drop= True)
```

```python
Healthcare_coverage_clean = Healthcare_coverage_clean.rename(columns = {'Reference area': 'Country', 'Frequency of observation': 'Annual Insurance', 'TIME_PERIOD': 'Year',})
Healthcare_coverage_clean.reset_index(drop= True)
```

```python
#Government/compulsory health insurance IS A GOVERNMENT THAT REQUIRES HEALTH INSURANCE BY ALL CIVILIANS
#Total voluntary health insurance IS PRIMARY INSURANCE
#Complementary voluntary health insurance IS SECONDARY INSURANCE
#Primary voluntary health insurance IS PRIMARY INSURANCE
#Duplicate voluntary health insurance IS SECONDARY INSURANCE
#Supplementary voluntary health insurance IS SECONDARY INSURANCE
#Public and primary voluntary health insurance IS GOVERNMENT SUBSIDIZED 
```

```python
Healthcare_coverage_clean['Insurance type'] = Healthcare_coverage_clean['Insurance type'].replace({
    'Government/compulsory health insurance': 'Government Primary',
    'Total voluntary health insurance': 'Primary',
    'Complementary voluntary health insurance': 'Secondary',
    'Primary voluntary health insurance': 'Primary',
    'Duplicate voluntary health insurance': 'Secondary',
    'Supplementary voluntary health insurance': 'Secondary',
    'Public and primary voluntary health insurance': 'Government Secondary'
})
Healthcare_coverage_clean.reset_index(drop= True)
```

# 1b) Tobacco consumption

```python
type(Tobacco_consumption)
```

```python
Tobacco_consumption.Measure.value_counts()
```

```python
Daily_Smokers = Tobacco_consumption.loc[Tobacco_consumption.Measure == '% of population aged 15+ who are daily smokers']
Daily_Smokers = Daily_Smokers.reset_index(drop = True)
Daily_Smokers = Daily_Smokers.rename(columns = {'Value': 'Percent of Smokers'})
Daily_Smokers
```

```python
type(Daily_Smokers)
```

# Daily Smokers tables

```python
#Alphabetical Country
pivot_df = Daily_Smokers.pivot_table(index='Country', columns='Year', values='Percent of Smokers')

plt.figure(figsize=(12, 10))
sns.heatmap(pivot_df, cmap='YlGnBu', annot=True, fmt=".1f", linewidths=.5)
plt.title('Percent of Daily Smokers age 15+')
plt.xlabel('Year')
plt.ylabel('Country')
plt.show();
```

```python
#Descending
pivot_df = Daily_Smokers.pivot_table(index='Country', columns='Year', values='Percent of Smokers')

pivot_df = pivot_df.sort_values(by=pivot_df.columns[-1], ascending=False)

plt.figure(figsize=(12, 10))
sns.heatmap(pivot_df, cmap='YlGnBu', annot=True, fmt=".1f", linewidths=.5)
plt.title('Percent of Daily Smokers age 15+')
plt.xlabel('Year')
plt.ylabel('Country')
plt.show();
```
